```python
# Punto 7

import numpy as np
import matplotlib.pyplot as plt
num_cuadrados =  100


x_c = np.linspace(-1, 1, num_cuadrados + 1)
y_c = np.linspace(-1, 1, num_cuadrados + 1)


fig, ax = plt.subplots()


for i in range(len(x_c) - 1):
    for j in range(len(y_c) - 1):

        rect = plt.Rectangle((x_c[i], y_c[j]), x_c[i+1] - x_c[i], y_c[j+1] - y_c[j], edgecolor='black', facecolor='none')
        ax.add_patch(rect)


ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_aspect('equal')

radio = 1
centro_x = 0
centro_y = 0

x = np.linspace(-radio, radio, 200)
y_pos = np.sqrt(radio**2 - x**2)
y_neg = -np.sqrt(radio**2 - x**2)

plt.plot(x, y_pos, 'r')
plt.plot(x, y_neg, 'r')
plt.grid(True)
plt.show()

def F(x,y):
  return np.sqrt(1-x**2-y**2+0j)

def promedio(x_c,y_c):
  suma = 0
  for i in range(len(x_c) - 1):
    for j in range(len(y_c) - 1):

      A=abs(F(x_c[i],y_c[j]))
      B=abs(F(x_c[i+1],y_c[j]))
      C=abs(F(x_c[i],y_c[j+1]))
      D=abs(F(x_c[i+1],y_c[j+1]))

      if np.sqrt((x_c[i])**2+(y_c[j])**2)>1:
        A=0

      if np.sqrt((x_c[i+1])**2+(y_c[j])**2)>1:
        B=0

      if np.sqrt((x_c[i])**2+(y_c[j+1])**2)>1:
        C=0

      if np.sqrt((x_c[i+1])**2+(y_c[j+1])**2)>1:
        D=0

      suma += ((A+B+C+D)/4)*(x_c[i+1]-x_c[i])*(y_c[j+1]-y_c[j])




  return suma




```

```python
# punto 17
from scipy.special import laguerre


def polinomio_laguerre(n):
  Polis=[]
  for i in range(n+1):
    Polis.append(laguerre(i))

  return Polis

def ceros_L(n):
  Ceros=[]
  for i in range(n+1):
    coeffs = polinomio_laguerre(i)[i].c
    roots = np.roots(coeffs)
    if roots.size == 0:
        Ceros.append([0])
    elif roots.size == 1:
        Ceros.append(roots.tolist())
    else:
        Ceros.append(roots.tolist())
  return Ceros[n]

def pesos_L(n):

    coeffs = polinomio_laguerre(n)[n].c
    ceros = np.roots(coeffs)
    L_nplus1 = laguerre(n+1)

    pesos = []
    for xi in ceros:
        Li = np.polyval(L_nplus1, xi)
        peso = xi / ((n+1)**2 * Li**2)
        pesos.append(peso)

    return pesos

def f(x):
    return (x**3) / (1 - np.exp(-x))

def cuadratura(n):

    ceros = ceros_L(n)
    pesos = pesos_L(n)

    integral = 0
    for i in range(len(ceros)):
        integral += pesos[i] * f(ceros[i])

    return integral
```

```python
I_exacto = (np.pi**4) / 15

n_values = list(range(2, 11))

errores_relativos = []

for n in n_values:
    integral_approx = cuadratura(n)
    error_relativo = integral_approx/I_exacto
    errores_relativos.append(error_relativo)

plt.scatter(n_values, errores_relativos, marker='o', linestyle='-', color='b', label='Laguerre quadrature accuracy')
plt.title('Error relativo vs. Número de puntos de evaluación (n)', fontsize=12)
plt.xlabel('Número de puntos de evaluación (n)', fontsize=10)
plt.ylabel('Error relativo', fontsize=10)
plt.grid(True)
plt.legend()
plt.show()
```
