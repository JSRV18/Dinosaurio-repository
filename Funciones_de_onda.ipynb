```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
phi0 = 0
x0 = 0.00000001
Dphi0 = 0.5

D = 1.44
C = 26.24
lphi = []
lx = []
def onda1(phi0, x0, Dphi0, E, dx):
  for i in range(10000):
    phi2 = -C*(E + D/x0)*phi0
    Dphi = Dphi0 + phi2*dx
    phi = phi0 + Dphi*dx
    lx.append(x0)
    lphi.append(phi0)
    Dphi0 = Dphi
    phi0 = phi
    x0 += dx

  return lx, lphi

xd = onda1(phi0, x0, Dphi0, -13.6027936, 0.0001)
plt.plot(xd[0], xd[1], c = 'teal', label = 'E = -13.6027936 eV')
plt.legend()
plt.ylabel('Ψ(x)')
plt.xlabel('x')
plt.title('Función de onda para n=1')
plt.show()
```
