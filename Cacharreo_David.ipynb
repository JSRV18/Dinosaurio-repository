```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')

file_path ="/content/drive/My Drive/Carbono.xlsx"
df = pd.read_excel(file_path)
```

```python
selected_columns = df.iloc[:, [0, 3]]
selected_columns
```

```python
import numpy as np
import matplotlib.pyplot as plt

anios =  np.array(df.iloc[:,0])
carbino = np.array(df.iloc[:,3])

plt.plot(anios, carbino, c="hotpink")
plt.title("Carbono vs Años")
plt.xlabel("Años")
plt.ylabel("Carbono")
plt.grid(True, c="lightgray")
plt.show()
```

```python
from scipy.optimize import curve_fit

def lineal (x, m, b):
  return m*x + b

params, _ = curve_fit(lineal, anios, carbino)

params

plt.plot(anios, carbino, c="pink")
plt.plot(anios, lineal(anios, *params), c="purple")
plt.title("Carbono vs Años")
plt.xlabel("Años")
plt.ylabel("Carbono")
plt.grid(True, c="lightgray")
```

```python
# Datos hijos

hijes = lineal(2035, *params)
hijes

nietes = lineal(2065, *params)
nietes

hijes, nietes
```
