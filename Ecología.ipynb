```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

import io
import pandas as pd
Abun1 = pd.read_csv(io.StringIO('''
"0,4347826087"
"0,1739130435"
"0,08695652174"
"0,2608695652"
0
"0,04347826087"
'''), header=None)

Abun1.sort_values(by=0, inplace=True, ascending=False)
Abun1.reset_index(drop=True, inplace=True)

index = np.array(Abun1.index) + 1
Abun = np.array(Abun1[0].str.replace(',', '.').astype(float)) *115
```

```python
import io
import pandas as pd
Abun2=pd.read_csv(io.StringIO('''
"0,24"
"0,32"
0
"0,2"
"0,16"
"0,08"
'''), header=None)

Abun3=pd.read_csv(io.StringIO('''
"0,125"
"0,15"
"0,175"
"0,225"
"0,15"
"0,175"
'''), header=None)

Abun2.sort_values(by=0, inplace=True, ascending=False)
Abun2.reset_index(drop=True, inplace=True)

index2 = np.array(Abun2.index) + 1
Abun2 = np.array(Abun2[0].str.replace(',', '.').astype(float))*125
plt.plot(index, Abun, label="Muestra 1", color="darkred")
plt.plot(index2, Abun2, label="Muestra 2", color="darkolivegreen")

plt.scatter(index, Abun, color="darkred")
plt.scatter(index2, Abun2, color="darkolivegreen")

Abun3.sort_values(by=0, inplace=True, ascending=False)
Abun3.reset_index(drop=True, inplace=True)

index3 = np.array(Abun3.index) + 1
Abun3 = np.array(Abun3[0].str.replace(',', '.').astype(float))*200

plt.plot(index3, Abun3, label="Muestra 3", color="teal")
plt.scatter(index3, Abun3, color="teal")
plt.title("Curvas de rango abundancia para cada muestra")
plt.xlabel("Rango")
plt.ylabel("Abundancia")
plt.legend()
plt.show()
```

```python
LnA1=np.log(Abun[:-1]/115)
LnA2=np.log(Abun2[:-1]/125)
LnA3=np.log(Abun3/200)

H1=-np.sum(LnA1*Abun[:-1]/115)
H2=-np.sum(LnA2*Abun2[:-1]/125)
H3=-np.sum(LnA3*Abun3/200)

np.round(H1,2), np.round(H2,2), np.round(H3,2)
```
